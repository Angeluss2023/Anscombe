```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Cargar el archivo CSV con la ruta completa
file_path = 'C:/Users/Bryam/Desktop/Mineria de datos/A1/df_anscombe.csv'
df = pd.read_csv(file_path)

# Crear una figura para los 4 grficos del conjunto de Anscombe
sns.set(style="whitegrid")

# Crear una figura con 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Grficas del 'quarteto de Anscombe'
for i, (name, group) in enumerate(df.groupby('group')):
    ax = axs[i // 2, i % 2]
    ax.scatter(group['x'], group['y'])
    ax.set_title(f'Group {name}')
    ax.set_xlim(0, 20)
    ax.set_ylim(0, 15)
    ax.set_xlabel('x')
    ax.set_ylabel('y')

plt.tight_layout()
plt.show()

# Crear los boxplots para las distribuciones de los grupos
plt.figure(figsize=(8, 6))
sns.boxplot(x='group', y='y', data=df)
plt.title("Boxplots por grupo")
plt.show()
```
